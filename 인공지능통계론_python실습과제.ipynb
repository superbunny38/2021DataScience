# Standardization

(p. 400~403)

image.png

```python
import pandas as pd
import numpy as np
height = [166, 186, 159, 174, 172, 148, 159, 171, 163, 165]
mean_height = np.mean(height)#평균
mean_adjusted = height - mean_height
sd_height = np.std(height)#표준편차
standardized = mean_adjusted/sd_height
```

```python
mean_height
```

```python
np.mean(standardized)#standardized된 데이터의 평균
```

```python
np.var(standardized)#standardized된 데이터의 분산
```

```python
np.std(standardized)#standardized된 데이터의 표준편차
```

```python
#데이터프레임 만들기
prep_height = pd.DataFrame({'height':height, 'mean_adjusted':mean_adjusted, 'standardized':standardized})
prep_height
```

```python
#데이터프레임 만들기
prep_height = pd.DataFrame({'height':height, 'mean_adjusted':mean_adjusted, 'standardized':standardized})
prep_height
```
